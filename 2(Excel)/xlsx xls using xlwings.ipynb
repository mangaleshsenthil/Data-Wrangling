```python
import xlwings as xw

with xw.App(visible=False) as app:
    
    print("--- Processing sample1.xlsx (Using xlwings) ---")
    
    wb = app.books.open('sample1.xlsx')
    sheet = wb.sheets[0]
    
    raw_data = sheet.range('A1').expand().value 
    
    headers = raw_data[0]
    rows = raw_data[1:]
    
    sales_data = [dict(zip(headers, row)) for row in rows]
    
    sales_data.sort(key=lambda x: x['Value'], reverse=True)
    
    print("\nTop 10 High Value Transactions:")
    for entry in sales_data[:10]:
        print(f"{entry['Sales_Rep_Name']} ({int(entry['Year'])}): {entry['Value']:.2f}")

    high_value_sales = [x for x in sales_data if x['Value'] > 50000]
    print(f"\nNumber of transactions > 50,000: {len(high_value_sales)}")

    new_wb = app.books.add()
    new_sheet = new_wb.sheets[0]
    new_sheet.name = "Analysis Output"
    
    output_rows = [headers] + [[row[h] for h in headers] for row in high_value_sales]
    
    new_sheet.range('A1').value = output_rows
    new_wb.save('output_analysis_xlwings.xlsx')
    new_wb.close()
    print("Saved filtered data to 'output_analysis_xlwings.xlsx'")

    sheet.range((2, 6)).value = "Verified by xlwings"
    sheet.range('F2:F6').value = [[("Verified by xlwings")] for _ in range(5)]
    
    wb.save('sample1_updated_xlwings.xlsx')
    wb.close()
    print("Saved updated file to 'sample1_updated_xlwings.xlsx'")

    print("\n\n--- Processing sample1.xls (Using xlwings) ---")
    
    wb_xls = app.books.open('sample1.xls')
    sheet_xls = wb_xls.sheets[0]
    
    raw_data_xls = sheet_xls.range('A1').expand().value
    headers_xls = raw_data_xls[0]
    rows_xls = raw_data_xls[1:]
    
    people_data = [dict(zip(headers_xls, row)) for row in rows_xls]
    
    people_data.sort(key=lambda x: x['Age'], reverse=True)
    
    print("\nTop 5 Oldest People:")
    for p in people_data[:5]:
        full_name = f"{p['First Name']} {p['Last Name']}"
        print(f"{full_name} ({p['Country']}): {int(p['Age'])}")

    adults = [p for p in people_data if p['Age'] > 30]
    print(f"\nNumber of people over 30: {len(adults)}")
    
    sheet_xls.range('H2').value = "Parsed by xlwings"
    
    wb_xls.save('sample1_updated_xlwings.xls')
    wb_xls.close()
    print("Saved updated legacy file to 'sample1_updated_xlwings.xls'")
```
