```python
from python_calamine import CalamineWorkbook
from openpyxl import Workbook
import xlwt

print("--- Processing sample1.xlsx (Using calamine) ---")

wb = CalamineWorkbook.from_path("sample1.xlsx")
rows = wb.get_sheet_by_index(0).to_python()

headers = rows[0]
sales_data = [dict(zip(headers, row)) for row in rows[1:]]

sales_data.sort(key=lambda x: x['Value'], reverse=True)

print("\nTop 10 High Value Transactions:")
for entry in sales_data[:10]:
    print(f"{entry['Sales_Rep_Name']} ({int(entry['Year'])}): {entry['Value']:.2f}")

high_value_sales = [x for x in sales_data if x['Value'] > 50000]
print(f"\nNumber of transactions > 50,000: {len(high_value_sales)}")

new_wb = Workbook()
new_ws = new_wb.active
new_ws.title = "Analysis Output"
new_ws.append(headers)

for item in high_value_sales:
    new_ws.append([item[h] for h in headers])

new_wb.save("output_analysis_calamine.xlsx")
print("Saved filtered data to 'output_analysis_calamine.xlsx'")

updated_rows = [list(headers) + ["Verification"]]
for i, row in enumerate(rows[1:]):
    verification = "Verified by calamine" if i < 5 else None
    updated_rows.append(list(row) + [verification])

update_wb = Workbook()
update_ws = update_wb.active
for row in updated_rows:
    update_ws.append(row)

update_wb.save("sample1_updated_calamine.xlsx")
print("Saved updated file to 'sample1_updated_calamine.xlsx'")


print("\n\n--- Processing sample1.xls (Using calamine) ---")

wb_xls = CalamineWorkbook.from_path("sample1.xls")
rows_xls = wb_xls.get_sheet_by_index(0).to_python()

headers_xls = rows_xls[0]
people_data = [dict(zip(headers_xls, row)) for row in rows_xls[1:]]

people_data.sort(key=lambda x: x['Age'], reverse=True)

print("\nTop 5 Oldest People:")
for p in people_data[:5]:
    full_name = f"{p['First Name']} {p['Last Name']}"
    print(f"{full_name} ({p['Country']}): {int(p['Age'])}")

adults = [p for p in people_data if p['Age'] > 30]
print(f"\nNumber of people over 30: {len(adults)}")

write_book = xlwt.Workbook()
write_sheet = write_book.add_sheet("Sheet1")

headers_updated = list(headers_xls) + ["Notes"]
for col_idx, header in enumerate(headers_updated):
    write_sheet.write(0, col_idx, header)

for row_idx, row_data in enumerate(rows_xls[1:], start=1):
    note = "Parsed by calamine" if row_idx == 1 else ""
    row_content = list(row_data) + [note]
    for col_idx, cell_value in enumerate(row_content):
        write_sheet.write(row_idx, col_idx, cell_value)

write_book.save("sample1_updated_calamine.xls")
print("Saved updated legacy file to 'sample1_updated_calamine.xls'")
```
