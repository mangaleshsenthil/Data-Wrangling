```python
import pandas as pd
# Processing sample1.xlsx 
print("Processing Using Pandas")
print("\n Processing sample1.xlsx")


# 1. Read the Excel file into a DataFrame
df_xlsx = pd.read_excel("sample1.xlsx")

# 2. Analysis: Sort Top 10 by 'Value'
top_10_sales = df_xlsx.sort_values(by="Value", ascending=False).head(10)

print("\nTop 10 High Value Transactions:")
print(top_10_sales[["Sales_Rep_Name", "Year", "Value"]].to_string(index=False))

# 3. Analysis: Filter Data (Value > 50,000)
high_value_df = df_xlsx[df_xlsx["Value"] > 50000]
print(f"\nNumber of transactions > 50,000: {len(high_value_df)}")

# 4. Write Filtered Data to a New File
high_value_df.to_excel("output_analysis_pandas.xlsx", index=False)
print("Saved filtered data to 'output_analysis_pandas.xlsx'")

# 5. Update Original Data
df_xlsx["Verification"] = None  # Create empty column
df_xlsx.loc[0:4, "Verification"] = "Verified by Pandas"

# Save the updated full dataset to a new file
df_xlsx.to_excel("sample1_updated_pandas.xlsx", index=False)
print("Saved updated file to 'sample1_updated_pandas.xlsx'")


print("\n Processing sample1.xls")

# 1. Read the Legacy Excel file
# explicitly use xlrd engine (installed above) to read .xls files
df_xls = pd.read_excel("sample1.xls", engine="xlrd")

# 2. Analysis: Sort Top 5 Oldest People
top_5_oldest = df_xls.sort_values(by="Age", ascending=False).head(5)

# Create a full name column for display
top_5_oldest["Full Name"] = top_5_oldest["First Name"] + " " + top_5_oldest["Last Name"]

print("\nTop 5 Oldest People:")
print(top_5_oldest[["Full Name", "Country", "Age"]].to_string(index=False))

# 3. Analysis: Filter Age > 30
adults_df = df_xls[df_xls["Age"] > 30]
print(f"\nNumber of people over 30: {len(adults_df)}")

# 4. Update Original Data
df_xls.loc[0, "Notes"] = "Parsed by Pandas"

# Save updated legacy file (xlwt engine is installed above)
df_xls.to_excel("sample1_updated_pandas.xls", index=False) 
print("Saved updated legacy file to 'sample1_updated_pandas.xls'")
```
