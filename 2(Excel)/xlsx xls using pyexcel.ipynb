```python
import pyexcel as pe

# Processing sample1.xlsx
print("--- Processing sample1.xlsx (Using pyexcel) ---")

# 1. Read the data
# get_records() automatically treats the first row as headers and returns a list of dictionaries
records_xlsx = pe.get_records(file_name="sample1.xlsx")

# 2. Analysis: Sort Top 10 by 'Value'
# Since it's a standard list of dicts, we use Python's sorted()
sorted_sales = sorted(records_xlsx, key=lambda x: x['Value'], reverse=True)

print("\nTop 10 High Value Transactions:")
for row in sorted_sales[:10]:
    print(f"{row['Sales_Rep_Name']} ({row['Year']}): {row['Value']}")

# 3. Analysis: Filter Data (Value > 50,000)
high_value_sales = [row for row in records_xlsx if row['Value'] > 50000]
print(f"\nNumber of transactions > 50,000: {len(high_value_sales)}")

# 4. Write Filtered Data to a New File
# pyexcel makes saving simple—just pass the list of dicts
pe.save_as(records=high_value_sales, dest_file_name="output_analysis_pyexcel.xlsx")
print("Saved filtered data to 'output_analysis_pyexcel.xlsx'")


# Processing sample1.xls 
print("\n\n--- Processing sample1.xls (Using pyexcel) ---")

# 1. Read the Legacy Excel file
# Note: pyexcel handles the .xls extension automatically without needing extra code
records_xls = pe.get_records(file_name="sample1.xls")

# 2. Analysis: Sort Top 5 Oldest People
sorted_people = sorted(records_xls, key=lambda x: x['Age'], reverse=True)

print("\nTop 5 Oldest People:")
for p in sorted_people[:5]:
    print(f"{p['First Name']} {p['Last Name']} ({p['Country']}): {p['Age']}")

# 3. Analysis: Filter Age > 30
adults = [p for p in records_xls if p['Age'] > 30]
print(f"\nNumber of people over 30: {len(adults)}")

# 4. Update Original Data (Non-Destructive)
# We can modify the list of dictionaries and save it back
for row in records_xls:
    row["Notes"] = "Parsed by pyexcel" # Adds a new key/column

# Save as new file
pe.save_as(records=records_xls, dest_file_name="sample1_updated_pyexcel.xls")
print("Saved updated legacy file to 'sample1_updated_pyexcel.xls'")
```
